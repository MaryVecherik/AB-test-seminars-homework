### 2. Шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов

Наибольшее количество балов набрала 2 гипотеза: если предложить 10% скидку на первую покупку для новых пользователей, то мы сможем повысить количество регистраций новых пользователей. 

Что мы делаем в каждой из групп: контрольная - без изменений, тестовая – даём новым пользователям код на скидку 10%. 

На каких пользователях мы тестируем: новые пользователи. 

Метрики: 
* Кол-во новых регистраций. 
* Доход от новых пользователей. 

Ожидаемый эффект: ожидается значительное увеличение регистраций новых пользователей, получивших код скидки, что приведет к увеличению дохода при минимальном влиянии на стоимость предлагаемой скидки.

План действий в зависимости от результата эксперимента: 
* Проведем A/B-тестирование, случайным образом разделив новых пользователей на две группы: одна группа получит код скидки 10% на свою первую покупку, а другая группа не получит никакой скидки.
* Измерим показатели для обеих групп и сравним результаты через месяц.

Избежать Peeking problem:
* Вычислим длительность теста и будем придерживаться сроков, которые мы посчитали.
* Будем мониторить промежуточные результаты эксперимента, но принимать решение будем только тогда, когда закончится экспермент.

### 3. Анализ выше указанных метрик

1. Соберём данные о регистрациях и доходах для контрольной и тестовой групп.
2. Разделим данные на выборки – каждая выборка будет содержать n пользователей, которые были направлены на каждую метрику.
3. Для каждой выборки будем считать количество новых регистраций и доход от новых пользователей для контрольной и тестовой групп.
4. С помощью python и его библиотек сгенерируем n выборок.
5. Для каждой выборки будем считать различие между контрольной и тестовой групп в количестве новых регистраций и доходе от новых пользователей.
6. Построим распределение полученных различий, оценим доверительные интервалы для обеих метрик.
7. Если доверительный интервал не содержит 0 (для метрики регистраций) или отрицательных значений (для метрики дохода), то можем сделать вывод о том, что эксперимент был успешным и можно внедрять гипотезу.
